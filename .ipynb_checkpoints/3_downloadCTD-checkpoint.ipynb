```python
import requests
import os
import pandas as pd
```

```python
# User defined variables
cruise = "EN657"
```

```python
# Create output directory
output_dir = rf"C:/Users/ocron/OneDrive - Massachusetts Institute of Technology/Documents/Research/Generals/Research_Report2/Data/IFCB/{cruise}/CTD/"
os.makedirs(output_dir, exist_ok=True)

meta_dir = rf"C:/Users/ocron/OneDrive - Massachusetts Institute of Technology/Documents/Research/Generals/Research_Report2/Data/IFCB/{cruise}/"
```

# Download CTD Metadata

```python
ctd_meta = f"https://nes-lter-data.whoi.edu/api/ctd/{cruise.lower()}/metadata.csv"

ctd_meta_filename = f"{cruise.lower()}_ctd_metadata.csv"  # or provide full path
filepath = os.path.join(meta_dir, ctd_meta_filename)

# Download and save
response = requests.get(ctd_meta)
response.raise_for_status()  # raise error if download fails

with open(filepath, 'wb') as f:
    f.write(response.content)

print(f"Saved to {filepath}")
```

# Download data from each CTD

```python
# Base URL format
base_url = f"https://nes-lter-data.whoi.edu/api/ctd/{cruise}/cast_{{}}.csv"
```

```python
failed_count = 0
max_consecutive_failures = 3
cast_num = 1

while failed_count < max_consecutive_failures:
    filename = f"{cruise}_cast_{cast_num}.csv"
    filepath = os.path.join(output_dir, filename)

    if os.path.exists(filepath):
        print(f"Skipping {filename} (already exists)")
        cast_num += 1
        failed_count = 0  # reset on successful skip
        continue

    url = base_url.format(cast_num)

    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(filepath, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {filename}")
        failed_count = 0  # reset on successful download
    except requests.exceptions.RequestException as e:
        print(f"Failed to download cast {cast_num}: {e}")
        failed_count += 1

    cast_num += 1
```
